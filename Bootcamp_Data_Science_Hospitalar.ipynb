### Desafio 04

Desenvolver uma função para a limpeza e tratamento dos dados, retornando duas tabelas "população" e "unida".

```
df_uf_recent = df_datasus_util['2020/Jul'].to_frame()
df_uf_recent.head()
```

```
def dados_pop_to_df(dados_populacao,target_month):

  from io import StringIO
  dados_populacao_io = StringIO(dados_populacao)
  dados_pop_csv = pd.read_csv(dados_populacao_io,sep="\t")
  dados_pop_csv = dados_pop_csv.drop(columns=['Posição','País comparável','% da pop. total'])
  dados_pop_csv = dados_pop_csv.rename(columns={'Unidade federativa':'UF','População':'populacao'})
  dados_pop_csv.index = dados_pop_csv['UF']
  dados_pop_csv = dados_pop_csv[1:]
  dados_pop_csv = dados_pop_csv.drop(columns=['UF'])
  dados_pop_csv['populacao'] = dados_pop_csv['populacao'].str.replace(' ','',regex=True).astype(int)

  
  df_uf_recent = df_datasus_util[target_month].to_frame()
  df_uf_recent.index = df_uf_recent.index.str[3:]
  df_uf_recent.index = df_uf_recent.index.str.strip()

  for estado in df_uf_recent.index:
    dados_pop_csv.index = dados_pop_csv.index.str.replace(f'{estado} {estado}',estado)
    dados_pop_csv.index = dados_pop_csv.index.str.strip()

  pop_e_gastos_recent = df_uf_recent.join(dados_pop_csv)
  pop_e_gastos_recent = pop_e_gastos_recent.rename(columns={target_month:'gastos_atuais'})
  pop_e_gastos_recent = pop_e_gastos_recent.sort_values('gastos_atuais', ascending = False)
  pop_e_gastos_recent['gasto_per_capita'] = pop_e_gastos_recent['gastos_atuais']/pop_e_gastos_recent['populacao']

  return pop_e_gastos_recent
```

```
df_uf_gasto_pop_07 = dados_pop_to_df(dados_populacao,'2020/Jul')
df_uf_gasto_pop_06 = dados_pop_to_df(dados_populacao,'2020/Jun')
```
